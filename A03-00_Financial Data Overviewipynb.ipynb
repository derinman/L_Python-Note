```python
data14 = pd.read_csv('data/tr_eikon_eod_data.csv',index_col=0, parse_dates=True)
window = 20
code = 'AAPL.O'
data14 = pd.DataFrame(data14[code])
data14['{}_min'.format(window)] = data14[code].rolling(window=window).min()
data14['{}_max'.format(window)] = data14[code].rolling(window=window).max()
data14['{}_median'.format(window)] = data14[code].rolling(window=window).median()
data14['{}_mean'.format(window)] = data14[code].rolling(window=window).mean()
data14['{}_ewma'.format(window)] = data14[code].ewm(halflife=0.5, min_periods=window).mean()
data14.dropna()
```

```python
data15 = pd.read_csv('data/tr_eikon_eod_data.csv',index_col=0, parse_dates=True)
window = 20
code = 'AAPL.O'

#dropna
data15 = pd.DataFrame(data15[code]).dropna()
data15['{}_min'.format(window)] = data15[code].rolling(window=window).min()
data15['{}_max'.format(window)] = data15[code].rolling(window=window).max()
data15['{}_median'.format(window)] = data15[code].rolling(window=window).median()
data15['{}_mean'.format(window)] = data15[code].rolling(window=window).mean()
#A02-06_Pandas Statistics Rolling.ipynb
data15['{}_ewma'.format('30half')] = data15[code].ewm(halflife=30, min_periods=60).mean()
data15['{}_ewma'.format('60half')] = data15[code].ewm(halflife=60, min_periods=120).mean()


ax = data15[[   '{}_min'.format(window),
                '{}_mean'.format(window),
                '{}_max'.format(window), 
                '{}_ewma'.format('30half'),
                '{}_ewma'.format('60half')]
            ].iloc[-200:].plot(figsize=(20, 14), style=['g--', 'r--', 'g--','y--','b--'], lw=0.8)  
data15[code].iloc[-200:].plot(ax=ax, lw=3.0)
```

```python
#simple moving average(SMA)
data16 = pd.read_csv('data/tr_eikon_eod_data.csv',index_col=0, parse_dates=True)

code = 'AAPL.O'
data16 = pd.DataFrame(data16[code]).dropna()
data16['SMA0'] = data16[code].rolling(window=21).mean()
data16['SMA1'] = data16[code].rolling(window=42).mean()
data16['SMA2'] = data16[code].rolling(window=252).mean()
data16['SMA3'] = data16[code].rolling(window=365).mean()
data16[[code,'SMA0', 'SMA1', 'SMA2','SMA3']].plot(figsize=(20, 12))  
```

```python
#simple moving average(SMA)
data17 = pd.read_csv('data/tr_eikon_eod_data.csv',index_col=0, parse_dates=True)

code = 'AAPL.O'
data17 = pd.DataFrame(data17[code]).dropna()
data17['SMA1'] = data17[code].rolling(window=42).mean()
data17['SMA2'] = data17[code].rolling(window=252).mean()
data17['positions'] = np.where(data17['SMA1'] > data17['SMA2'],  1,  -1)  

ax = data17[[code, 'SMA1', 'SMA2', 'positions']].plot(figsize=(20, 12),secondary_y='positions')
ax.get_legend().set_bbox_to_anchor((0.12, 0.9));
```

```python
data22 = pd.read_csv('data/tr_eikon_eod_data.csv', index_col=0, parse_dates=True)
data22 = data22[['.SPX', '.VIX']].dropna()
data22 = np.log(data22 / data22.shift(1)).dropna()

z1 = np.polyfit(data22['.SPX'], data22['.VIX'], 1)
f1 = np.poly1d(z1)

z2 = np.polyfit(data22['.SPX'], data22['.VIX'], 2)
f2 = np.poly1d(z2)

z3 = np.polyfit(data22['.SPX'], data22['.VIX'], 3)
f3 = np.poly1d(z3)

z6 = np.polyfit(data22['.SPX'], data22['.VIX'], 6)
f6 = np.poly1d(z6)

tmpX = np.linspace(data22['.SPX'].min(), data22['.SPX'].max(), 50)


ax = data22.plot(kind='scatter', x='.SPX', y='.VIX', figsize=(20, 12))
#ax = (data22['.SPX'],data22['.VIX'])
ax.scatter(tmpX, f1(tmpX),label='f1')
ax.scatter(tmpX, f2(tmpX),label='f1')
ax.scatter(tmpX, f3(tmpX),label='f2')
ax.scatter(tmpX, f6(tmpX),label='f6')
ax.legend(loc=2)
```
