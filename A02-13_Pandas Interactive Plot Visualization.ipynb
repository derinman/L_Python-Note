```python
import numpy as np
import pandas as pd
import cufflinks as cf
import plotly.offline as plyo  
```

```python
a = np.random.standard_normal((2500, 5)).cumsum(axis=0) 
#營業日，周一至周五
index = pd.date_range('2019-1-1', freq='B', periods=len(a)) 
df = pd.DataFrame(100 + 5 * a, columns=list('abcde'), index=index) 

plyo.iplot(df.iplot(asFigure=True), image='png', filename='ply_01')
```

```python
a = np.random.standard_normal((2500, 5)).cumsum(axis=0) 
#營業日，周一至周五
index = pd.date_range('2019-1-1', freq='B', periods=len(a)) 
df2 = pd.DataFrame(100 + 5 * a, columns=list('abcde'), index=index) 

plyo.iplot(
    df2[['a', 'b']].
    iplot(  asFigure=True,
            theme='polar',  
            title='A Time Series Plot',  
            xTitle='date',  
            yTitle='value',  
            mode={'a': 'markers', 'b': 'lines+markers'},  
            symbol={'a': 'circle', 'b': 'star'},  
            size=3.5,  
            colors={'a': 'blue', 'b': 'red'},  
        ),
    image='png',
    filename='ply_02'
)
```

```python
raw = pd.read_csv('./data/fxcm_eur_usd_eod_data.csv',
                    index_col=0, parse_dates=True) 
quotes = raw[['AskOpen', 'AskHigh', 'AskLow', 'AskClose']]  
quotes = quotes.iloc[-60:]  

qf = cf.QuantFig(
                quotes,  
                title='EUR/USD Exchange Rate',  
                legend='top',  
                name='EUR/USD'  
)
#布林通道
qf.add_bollinger_bands( periods=15,  
                        boll_std=2)  
#相對強弱指標
qf.add_rsi( periods=14,  
            showbands=False)  

plyo.iplot(
    qf.iplot(asFigure=True),
    image='png',
    filename='qf_01'
)
```
