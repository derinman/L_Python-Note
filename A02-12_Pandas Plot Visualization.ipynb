```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
np.random.seed(seed=123456)
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))
ts = ts.cumsum()

df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list("ABCD"))
df = df.cumsum()

ts.plot()
df.plot()
```

```python
#指定 x y

df2 = pd.DataFrame(np.random.randn(1000, 2), columns=["B", "C"]).cumsum()
df2['A'] = pd.Series(list(range(len(df2))))
df2.plot(x='A', y='B')
```

```python
#line style

t = np.arange(0., 5., 0.2)
legend_labels = ['Solid', 'Dashed', 'Dotted', 
                 'Dot-dashed', 'Points']
df3 = pd.DataFrame({0 : t,
                           1 : t**1.5,
                           2 : t**2.0,
                           3 : t**2.5,
                           4 : t**3.0})
ax = df3.plot(style=['r-', 'g--', 'b:', 'm-.', 'k:'])
ax.legend(legend_labels, loc='upper left')
```

```python
from matplotlib.dates import WeekdayLocator, DateFormatter, MonthLocator

s = pd.Series(np.random.randn(1096),index=pd.date_range('2012-01-01', '2014-12-31'))
walk_ts = s.cumsum()
walk_df = pd.DataFrame(walk_ts)

# plot Jan-Feb 2014
ax = walk_df.loc['2014-01':'2014-02'].plot()

# do the minor labels
weekday_locator = WeekdayLocator(byweekday=(0), interval=1)
ax.xaxis.set_minor_locator(weekday_locator)
ax.xaxis.set_minor_formatter(DateFormatter("%d\n%a"))

# do the major labels
ax.xaxis.set_major_locator(MonthLocator())
ax.xaxis.set_major_formatter(DateFormatter('\n\n\n%b\n%Y'))

ax.xaxis.grid(True, "minor") # turn on minor tick grid lines
ax.xaxis.grid(False, "major") # turn off major tick grid lines
```

```python
ax6 = plt.subplot2grid((2,1), (0,0))
ax7 = plt.subplot2grid((2,1), (1,0))
walk_df[walk_df.columns[0]].plot(ax = ax6)
walk_df[walk_df.columns[0]].plot(ax = ax7)
```
